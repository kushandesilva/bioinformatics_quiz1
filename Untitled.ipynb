```python
def LetterToNumber(Letter):
    if Letter == 'A':
        return 0
    elif Letter == 'C':
        return 1
    elif Letter == 'G':
        return 2
    elif Letter == 'T':
        return 3
    else:
        return "INVALID"

def PatternToIndex(Pattern):
    Base10 = 0
    PatternInput = list(Pattern)
    PatternToBase4 = []
    for Letter in PatternInput:
        PatternToBase4.append(LetterToNumber(Letter))
    PatternToBase4.reverse()
    for index, value in enumerate(PatternToBase4):
        Base10 += value * 4**index
    return Base10

print(PatternToIndex('CCAGA'))


"""
Code Challenge: Implement ComputingFrequencies to generate a frequency array.
    Input: A DNA string Text followed by an integer k.
    Output: FrequencyArray(Text).
"""
def ComputingFrequencies(Text, k):
    frequency = []
    for i in range(0, 4**k):
        frequency.append(0)
    for i in range(0, (len(Text) - (k-1))):
        currentPattern = Text[i:i+k]
        index = PatternToIndex(currentPattern)
        frequency[index] += 1
    result = ' '.join(str(number) for number in frequency)
    return result


print(ComputingFrequencies('CCAATGTCACTCATGCCACCTCCGCAGATCCGACCTTGTCCAGTAATTTCCGCTGTTTGTGGGCCGTTGCACAGAGTACTCAGTTATAACCGATGGACAGATGTTACGACTGGTTCCGCAGATATTCAGCTTGCGGAGCTGGAAATGACAAACTGGAGACATCCGTGCATACGTTACAGAGAAGACCTCTTTCATAGCACCCTCGCGGGAAGGTCGGGATTATCGGGCTGTATTAGTCCCTTGGGTTGGATTTCGAAGATGGTTCGCGACATGTCTGCGAAGGAGACCACACATCCAGGGAGTTACCGAACCGATAATGATTCGTGAGTTCGCTAAACCGCAATGCGAAGTGCGGGCTGCAGTAGATAAAGACGGATCCCGCACCTTTCGCCGCCATCGCCAGCCTGGTAGAGTGATACCAATACGCACTGCTTCGTTCAGTGCGAACTCATGGTCAGCCTCCTGCTTTGATGTGATAGACTCCAGGTATCTTTGGTGACGCGCACTACTCATAGGAAATTTAGCGTGCGACCACGTTGCAACCTATGCTACCAGCAGGCTTACTTGTGAAATAATTAAGTGCCCCAGATATATCTGAAGTAATCTCCTGTTGTGCAGACTGGGACTGAGACAGGAAAG', 6))
```
